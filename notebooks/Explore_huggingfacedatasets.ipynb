```python
def transform_dataset(ds, ds_name):
    ds["en"] = ds.apply(lambda x: x["translation"]["en"], axis=1)
    ds["hu"] = ds.apply(lambda x: x["translation"]["hu"], axis=1)
    #ds["dataset"] = ds_name

    ds = ds.drop(columns=["id", "translation"])
    
    return ds
```
